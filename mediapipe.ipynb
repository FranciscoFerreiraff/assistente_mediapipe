```python
import cv2
import mediapipe as mp
from mediapipe.framework.formats import landmark_pb2
import time
from datetime import datetime, timedelta
import webbrowser
import func
import os
```

```python
#importando os itens importantes
mp_drawing = mp.solutions.drawing_utils
mp_holistic = mp.solutions.holistic #existem variações do holistic(que pega mãos, face e outras partes do corpo)
#capturar imagem webcam
cap = cv2.VideoCapture(0) # 0 é o número da webcam, caria de pc para pc
#iniciar o modelo holistic
#chamando e setando a confiança, aumente se quiser mais precisão
with mp_holistic.Holistic(min_detection_confidence = 0.5, min_tracking_confidence=0.5) as holistic:

    while cap.isOpened():
        ret,frame = cap.read() #lendo o feed da webcam

        #recolorir o feed
        image = cv2.cvtColor(frame, cv2.COLOR_BGR2RGB)
        #fazer detecções
        results = holistic.process(image)
        
        #recolorir a imagem de volta para BGR para renderizar no opencv(só aceita RBG)
        image = cv2.cvtColor(image, cv2.COLOR_RGB2BGR)
        #desenha os landmarks na face
        mp_drawing.draw_landmarks(image, results.face_landmarks, mp_holistic.FACE_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color = (80,110,10), thickness = 1, circle_radius = 4), #mudando cor do landmark
                                 mp_drawing.DrawingSpec(color = (80,256,121), thickness = 1, circle_radius = 2) #mudando cor da linha
                                 )

        #Mão direita
        mp_drawing.draw_landmarks(image, results.right_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color = (168,54,50), thickness = 2, circle_radius = 4), #mudando cor do landmark
                                 mp_drawing.DrawingSpec(color = (36,25,25), thickness = 2, circle_radius = 2) #mudando cor da linha
                                 )
        
        #Mão esquerda
        mp_drawing.draw_landmarks(image, results.left_hand_landmarks, mp_holistic.HAND_CONNECTIONS,
                                 mp_drawing.DrawingSpec(color = (168,54,50), thickness = 2, circle_radius = 4), #mudando cor do landmark
                                 mp_drawing.DrawingSpec(color = (36,25,25), thickness = 2, circle_radius = 2) #mudando cor da linha
                                 )        
        #pose
        #mp_drawing.draw_landmarks(image, results.pose_landmarks, mp_holistic.POSE_CONNECTIONS,
                                 #mp_drawing.DrawingSpec(color = (245,117,66), thickness = 2, circle_radius = 4), #mudando cor do landmark
                                 #mp_drawing.DrawingSpec(color = (245,66,230), thickness = 2, circle_radius = 2) #mudando cor da linha
                                 #)
        #só vai rodar se houver algo para capturar            
        if type(results.left_hand_landmarks) is type(None) or type(results.face_landmarks) is type(None) :
            pass
        #capturando pontos
        else:
            keypoints_left_hand = []
            keypoint_face = []
            for data_point in results.left_hand_landmarks.landmark:
                keypoints_left_hand.append({
                                     'X': data_point.x,
                                     'Y': data_point.y,
                                     'Z': data_point.z,
                                     'Visibility': data_point.visibility,
                                     })
            for data_point in results.face_landmarks.landmark:
                keypoint_face.append({
                                     'X': data_point.x,
                                     'Y': data_point.y,
                                     'Z': data_point.z,
                                     'Visibility': data_point.visibility,
                                     })
            if int(datetime.now().strftime('%S')) % 1 == 0:
                try:
                    #para correspondência ver a imagem do inicío
                    if abs(keypoints_left_hand[4]["X"] - keypoints_left_hand[8]["X"]) < 0.004: #sinal "ok"
                        #print("yeahh " + datetime.now().strftime('%S.%f'))
                        func.speak("Abrindo youtube")
                        webbrowser.open("http://youtube.com", new=1)
                        print("aa")
                        time.sleep(1.5)
                    elif abs(keypoints_left_hand[8]["X"] - keypoint_face[454]["X"]) < 0.01: #sinal "ouvindo"
                        #print("noooo " + datetime.now().strftime('%S.%f'))
                        func.speak("Estou ouvindo")
                        said = func.ouvir()
                        time.sleep(1.5)
                        if ("sports" or "sport") in said.lower():
                            func.speak("Abrindo Globo esporte")
                            webbrowser.open("https://ge.globo.com/", new=1)
                        if ("youtube") in said.lower():
                            func.speak("Abrindo Youtube")
                            webbrowser.open("http://youtube.com", new=1)
                        if "resident" in said.lower():
                            os.startfile(r"C:\Users\Franc\OneDrive\Área de Trabalho\jogos\Resident Evil 3.lnk")
                        if "excel" in said.lower():    
                            os.startfile(r"C:\ProgramData\Microsoft\Windows\Start Menu\Programs\Excel.lnk")
                except:
                    pass
      
        cv2.imshow("Webcam Feed", image)

        if cv2.waitKey(10) & 0xFF == ord("q"): #para parar aperte q
            break
#parar garantir que ao sair do loop ele vai fechar
cap.release()
cv2.destroyAllWindows()
```
